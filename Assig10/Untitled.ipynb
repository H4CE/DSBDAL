```python
import nltk 
import spacy
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize

sentences1=sent_tokenize(corpus1)
sentences2=sent_tokenize(corpus2)

print(sentences1)
print(sentences2)

word1=word_tokenize(corpus1)
word2=word_tokenize(corpus2)

print(word1)
print(word2)
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
nlp1_corpus=nlp(corpus1)
```

```python
for x in nlp1_corpus:
    print(x,"-->",spacy.explain(x.pos_),x.pos_)
```

```python
filtered_corpus1=[]
for word in nlp1_corpus:
    if not word.is_stop and not word.is_punct:
        filtered_corpus1.append(word.text)
```

```python
from nltk.stem import PorterStemmer

stemmer=PorterStemmer()

for word in filtered_corpus1:
    print(word,stemmer.stem(word))
```

```python
for word in nlp1_corpus:
    print(word.text,word.lemma_)
```
