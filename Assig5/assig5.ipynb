```python
import pandas as pd
df=pd.read_csv("nba.csv")
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
#for showing the entire rows and entire columns
```

```python
df.isnull().sum()
```

```python
#filling the null values of college column with previous value
df["College"]=df["College"].ffill()
```

```python
#filling the null values of salary column with median value
medianSalary=df["Salary"].median()
df["Salary"].fillna(medianSalary,inplace=True)
```

```python
#dropping the row with all columns null except the last two
df.dropna(how="any",axis=0,inplace=True)
```

```python
# providing the summary statistics for the numerical variable age, grouped by the categorical value "College"
df.groupby("College")["Age"].mean()
```

```python
# providing the summary statistics for the numerical variable salary, grouped by the categorical value [team and position]

df.groupby(["Team","Position"])["Salary"].mean()
```

```python
#defining a custom age group and then find out the numerical values grouped by age groups
#for that, first lets see the unique values present in the age column
ages=df["Age"].unique()
ages.sort()
print(ages)
```

```python
#bin array for setting criteria to slice by while using the cut method:
bins=[19,26,31,36,40]
age_group_labels=["19-25","26-30","31-35","36-40"]
df["AgeGroups"]=pd.cut(df["Age"],bins=bins,labels=age_group_labels)
```

```python
df.groupby("AgeGroups")["Salary"].describe()
```
